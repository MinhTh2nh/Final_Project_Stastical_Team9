```python
HAG=pd.read_csv(r"/content/HAGLdataset (4).csv")
HAG
```

```python
# Convert the 'date' column to a datetime data type
#set the date as the index
# HPG=HPG.set_index(pd.DatetimeIndex(HPG['date'].values))
HAG['Time'] = np.arange(len(HAG.index))
HAG['price'].plot(figsize=(16,5))
```

```python
HAG_2 = HAG.copy()
#plot the linear regression adn the close price
```

#create the new column called Numbers

```python
HAG_2['Numbers'] = list(range(0, len(HAG_2)))
#store the 'Numbers' to X and ClosePrice to Y
X = np.array(HAG_2[['Numbers']])
Y = HAG_2['price'].values
linearmodel = LinearRegression().fit(X,Y)
print('Intercept' , linearmodel.intercept_)
print('Slope' , linearmodel.coef_)
# HPG_2['Numbers']
```

```python
#prepare the data for visualization
#store the predicted prices from the model and store them into y_pred
#y = ax + b 
y_pred = linearmodel.coef_* X + linearmodel.intercept_
#store the y_pred into new colums call Predicted
HAG_2['Predicted'] =  y_pred
#plot the data
HAG_2['Predicted'].plot(figsize=(10,5),legend = True ,color = 'orange')
HAG_2['price'].plot(figsize=(10,5),legend = True )
plt.title('Close Price History HAG')
```

```python
#độ phù hợp với tập test
#Độ phù hợp với mô hình (cho biết mô hình này phù hợp ở mức %) 
# #Train data
#train model bằng tập HPG_2
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score
mae = mean_absolute_error(HAG_2['price'], HAG_2['Predicted'])
mape = mean_absolute_percentage_error(HAG_2['price'], HAG_2['Predicted'])
mse = mean_squared_error(HAG_2['price'], HAG_2['Predicted'])
rmse = np.sqrt(mse)
r2 = r2_score(HAG_2['price'], HAG_2['Predicted'])

print(f"MAE: {mae:.2f}")
print(f"MAPE: {mape * 100:.2f}%")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R2: {r2:.2f}")
```
