```python
#Create Time column has value is index
df['So'] = np.arange(len(df.index))
df
```

```python
#
df['newdate'] = pd.to_datetime(df.date)
type(df['newdate'])
```

```python
df.index = df['newdate']
df
```

```python
##visualization FPT stock price from 2015 to 2022
df.plot(x = 'So', y = 'price', figsize = (12,5))
```

```python
x = np.array(df.So, dtype = float)
y = np.array(df.price, dtype = float)
z = [1812,1813,1814,1815,1816,1817,1818,1819,1820,1821,1822,1823,1824,1825,1826,1827,1828,1829,1830,1831,1832,1833,1834,1835,1836,1837,1838,1839,1840,1841,1842]
```

```python
no_of_days = 30
future_day = pd.date_range(start= "11/30/2022",end = "12/30/2022")
```

```python
coefs = poly.polyfit(x, y, 9)
x_new = np.linspace(x[0], x[-1] + no_of_days, num = len(x) + no_of_days)
fit = poly.polyval(x_new, coefs)
pred = poly.polyval(z, coefs)
predictions = pd.DataFrame(future_day ,pred)
print(predictions)
```

```python
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error

fit = fit[30:].reshape(1812,)

mape = mean_absolute_percentage_error(y, fit)
mse = mean_squared_error(y, fit)
rmse = np.sqrt(mse)


print(f"MAPE: {mape * 100:.2f}%")
print(f"RMSE: {rmse:.2f}")
```
